```python
datasets=pd.read_excel(r'ObesityDataSet_raw_and_data_sinthetic.xlsx')
```

```python
datasets
```

```python
from sklearn.preprocessing import LabelEncoder
cat_col = datasets.select_dtypes(['object']).columns
datasets[cat_col] = datasets[cat_col].apply(lambda x: x.astype('category'))
datasets[cat_col] = datasets[cat_col].apply(lambda x: x.cat.codes)
```

```python
datasets
```

```python
y=datasets.iloc[:,-1:]
x=datasets.iloc[:,:-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
clf=SVC(decision_function_shape='ovr')
clf.fit(x_train,y_train)
```

```python
y_predict=clf.predict(x_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```
