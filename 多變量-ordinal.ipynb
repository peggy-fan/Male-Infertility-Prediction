```python

import numpy as np
import pandas as pd
from statsmodels.miscmodels.ordinal_model import OrderedModel
```

```python
datasets=pd.read_excel(r'ObesityDataSet_raw_and_data_sinthetic.xlsx')
```

```python
datasets
```

```python
datasets['NObeyesdad'].dtype
```

```python
from sklearn.preprocessing import LabelEncoder
cat_col = datasets.select_dtypes(['object']).columns
datasets[cat_col] = datasets[cat_col].apply(lambda x: x.astype('category'))
datasets[cat_col] = datasets[cat_col].apply(lambda x: x.cat.codes)
```

```python
mod_log=OrderedModel(datasets['NObeyesdad'],
            datasets[['FAVC','FCVC','NCP','CAEC','CH2O','SCC','FAF','TUE','CALC','MTRANS']],
            distr='logit')

res_log =mod_log.fit(method='bfgs', disp=False)
res_log.summary()
```

```python
predicted = res_log.model.predict(res_log.params, exog=datasets[['FAVC','FCVC','NCP','CAEC','CH2O','SCC','FAF','TUE','CALC','MTRANS']])
predicted
```

```python
pred_choice = predicted.argmax(1)
print('Fraction of correct choice predictions')
print((np.asarray(datasets['NObeyesdad'].values) == pred_choice).mean())
```
